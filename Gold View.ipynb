```sql
USE CATALOG ushealthcaredynamics;
USE SCHEMA gold;
```

```sql
CREATE MATERIALIZED VIEW HospitalLookup AS
SELECT 
    dim_location_pk,
    location_name
FROM 
    ushealthcaredynamics.silver.hospital_lookup;

```

```sql
CREATE MATERIALIZED VIEW CptCode_lookup AS
SELECT
    dim_cpt_code_pk,
    cpt_desc,
    cpt_code,
    cpt_grouping,
    short_desc,
    is_cpt_code_missing
FROM
    ushealthcaredynamics.silver.cptcode_lookup;
```

```sql
CREATE MATERIALIZED VIEW DimDate AS
SELECT
    
    Date,
    date_year,
    date_day_name,
    date_month_name,
    date_day_of_month,
   date_month_number
   
  
FROM
    ushealthcaredynamics.silver.dimdate;
```

```sql
CREATE MATERIALIZED VIEW Physcian_Lookup AS
SELECT
    dim_physician_pk,
    provider_name,
    provider_fte,
    provider_npi,
    speciality_code
FROM
    ushealthcaredynamics.silver.physician_lookup;
```

```sql
CREATE MATERIALIZED VIEW Payer_Lookup AS
SELECT
    dim_payer_pk,
    payer_name
FROM
    ushealthcaredynamics.silver.payer_lookup;
```

```sql
CREATE MATERIALIZED VIEW DiagnosisCode_Lookup AS
SELECT
    dimDiagnosisCodePK,
    DiagnosisCode,
    DiagnosisCodeDescription,
    DiagnosisCodeGroup
FROM
    ushealthcaredynamics.silver.diagnosiscode_lookup;
```

```sql
CREATE MATERIALIZED VIEW Transaction_lookup AS
SELECT
    dim_transaction_pk,
    adjustment_reason,
    transaction,
    transaction_type
FROM
    ushealthcaredynamics.silver.transaction_lookup;
```

```sql
CREATE MATERIALIZED VIEW FactTable_lookup AS
SELECT *
FROM ushealthcaredynamics.silver.facttable
```
